```python
## 트레인 파일을 읽어서 실험용  set를 먼저 만든다.

import csv
import os
import pandas as pd


data_path = "/home/soon2soon/Notebooks/DreamChallenge/sc2/data/"
# processed_dataPath = "/home/soon2soon/Notebooks/DreamChallenge/sc2/processed_data/"
processed_dataPath = "/home/soon2soon/Notebooks/DreamChallenge/sc2/eval_processed_data/"


merged_tablePath_train = data_path + "id_file_mapping_train.csv"
merged_tablePath_test = data_path + "id_file_mapping_test.csv"


merged_table_train = pd.read_csv(merged_tablePath_train)
merged_table_test = pd.read_csv(merged_tablePath_test)

merged_table_train = merged_table_train.drop(merged_table_train.columns[0], axis=1)
merged_table_test = merged_table_test.drop(merged_table_test.columns[0], axis=1)

print(merged_table_train.shape)
print(merged_table_test.shape)


total_rows_train=len(merged_table_train.axes[0])
total_rows_test=len(merged_table_test.axes[0])

print("train set size:", total_rows_train)
print("test set size:", total_rows_test)
```

```python
## split trainset into two subset for training

train_subset = pd.DataFrame()
test_subset = pd.DataFrame()

train_split_size = int(merged_table_train.shape[0] * 0.9)

# print(train_split_size)
rest = merged_table_train.shape[0] - train_split_size
# print(rest)

train_subset = merged_table_train[:train_split_size]
test_subset = merged_table_train[train_split_size:]

```

```python
###############
mode = 'test'
###############


targetTable = pd.DataFrame()

# if mode == 'train':
#     targetTable = train_subset
# else:
#     targetTable = test_subset
    
    
if mode == 'train':
    targetTable = merged_table_train
else:
    targetTable = merged_table_test


file_acc_x = open(processed_dataPath + '/' +mode+ '/'+ "body_acc_x_"+mode+".txt", "w")
file_acc_y = open(processed_dataPath + '/' +mode+ '/'+ "body_acc_y_"+mode+".txt", "w")
file_acc_z = open(processed_dataPath + '/' +mode+ '/'+ "body_acc_z_"+mode+".txt", "w")

file_y_label = open(processed_dataPath + '/' +mode+ '/'+ "y_"+mode+".txt", "w")

## tracking each row's identity
file_mapping_label = open(processed_dataPath + '/' +mode+ '/'+ "mapping_"+mode+".txt", "w")


# phenotype : tremorScore  dyskinesiaScore  bradykinesiaScore
## 한줄에 파일 하나. 읽어서 데이터 처리. 
for index,row in targetTable.iterrows():
#     print(row['dataFileHandleId'], row['FILE_PATH'])
    
    with open(row['FILE_PATH']) as data_file:
        
        tsvin = csv.reader(data_file, delimiter='\t')

        #파일별로 초기화 되어야함
        acc_x = []
        acc_y = []
        acc_z = []
        
        idx = 0
        ## 여기서 타임스탭별로 데이터 읽어서 데이터 한 줄 만들고,,
        for row_t in tsvin:
            idx = idx + 1
            
            # skip header
            if idx == 1:
                continue
                
            
            ## x=1, y=2, z=3, magn=4
#             print(row_t[1])
            
            if idx <= 500:
                acc_x.append(row_t[1])
                acc_y.append(row_t[2])
                acc_z.append(row_t[3])
        
        y = -1
        
        
        if row['dyskinesiaScore'] == 0.0 or row['dyskinesiaScore'] == 1.0:
            y = 5
        else:
#             y = int(row['tremorScore'])
            y = int('0')
            
#         print(row['tremorScore'], row['dyskinesiaScore'], row['bradykinesiaScore'], y)
                        
        
        ## 파일에 쓴다

            
        for i in range(0, len(acc_x)):
            file_acc_x.write(str(float(acc_x[i])))
            file_acc_x.write(' ')
            
            file_acc_y.write(str(float(acc_y[i])))
            file_acc_y.write(' ')
            
            file_acc_z.write(str(float(acc_z[i])))
            file_acc_z.write(' ')
            
        file_y_label.write(str(y))
        file_mapping_label.write(str(row['dataFileHandleId']))
        
        if i < len(acc_x):
            file_acc_x.write('\n')
            file_acc_y.write('\n')
            file_acc_z.write('\n')
            file_y_label.write('\n')
            file_mapping_label.write('\n')
        
        
#         print(index, len(acc_x), row['tremorScore'], row['dyskinesiaScore'], row['bradykinesiaScore'], y)
#         print(idx)

file_acc_x.close()
file_acc_y.close()
file_acc_z.close()
file_y_label.close()
file_mapping_label.close()

print(mode, "File write done.")
```
