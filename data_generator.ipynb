## Data generator

```python

        

acc_x = []
acc_y = []
acc_z = []

rotation_x = []
rotation_y = []
rotation_z = []

attitude_x = []
attitude_y = []
attitude_z = []

gravity_x = []
gravity_y = []
gravity_z = []



mode = "train"

        

processed_data_path = "/home/soon2soon/Notebooks/DreamChallenge/data/processed_data_additional"


file_acc_x = open(processed_data_path + '/' +mode+ '/'+ "body_acc_x_"+mode+".txt", "w")
file_acc_y = open(processed_data_path + '/' +mode+ '/'+ "body_acc_y_"+mode+".txt", "w")
file_acc_z = open(processed_data_path + '/' +mode+ '/'+ "body_acc_z_"+mode+".txt", "w")

file_rotation_x = open(processed_data_path + '/' +mode+ '/'+ "body_rotation_x_"+mode+".txt", "w")
file_rotation_y = open(processed_data_path + '/' +mode+ '/'+ "body_rotation_y_"+mode+".txt", "w")
file_rotation_z = open(processed_data_path + '/' +mode+ '/'+ "body_rotation_z_"+mode+".txt", "w")

file_attitude_x = open(processed_data_path + '/' +mode+ '/'+ "body_attitude_x_"+mode+".txt", "w")
file_attitude_y = open(processed_data_path + '/' +mode+ '/'+ "body_attitude_y_"+mode+".txt", "w")
file_attitude_z = open(processed_data_path + '/' +mode+ '/'+ "body_attitude_z_"+mode+".txt", "w")

file_gravity_x = open(processed_data_path + '/' +mode+ '/'+ "body_gravity_x_"+mode+".txt", "w")
file_gravity_y = open(processed_data_path + '/' +mode+ '/'+ "body_gravity_y_"+mode+".txt", "w")
file_gravity_z = open(processed_data_path + '/' +mode+ '/'+ "body_gravity_z_"+mode+".txt", "w")


## is this person patient or normal?
file_y_label = open(processed_data_path + '/' +mode+ '/'+ "y_"+mode+".txt", "w")

file_mapping_label = open(processed_data_path + '/' +mode+ '/'+ "mapping_"+mode+".txt", "w")
    
    
target_arr = []


if mode == "train":
    target_arr = pathList_train
else:
    target_arr = pathList_test
    
print(len(target_arr))
   
    
    
for index in range (0, len(target_arr)):
    with open(target_arr[index]) as data_file:    
#         print(target_arr[index])
        data = json.load(data_file)
#         print("len(data)", len(data))
        
        # reset data array
        acc_x = []
        acc_y = []
        acc_z = []

        rotation_x = []
        rotation_y = []
        rotation_z = []

        attitude_x = []
        attitude_y = []
        attitude_z = []
        
        gravity_x = []
        gravity_y = []
        gravity_z = []
        
        for listIndex in range(0, len(data)):
            ### acc
            acc_x.append(data[listIndex]['userAcceleration']['x'])
            acc_y.append(data[listIndex]['userAcceleration']['y'])
            acc_z.append(data[listIndex]['userAcceleration']['z'])
            
            ### rotationRate
            rotation_x.append(data[listIndex]['rotationRate']['x'])
            rotation_y.append(data[listIndex]['rotationRate']['y'])
            rotation_z.append(data[listIndex]['rotationRate']['z'])
            
            ### attitude
            attitude_x.append(data[listIndex]['attitude']['x'])
            attitude_y.append(data[listIndex]['attitude']['y'])
            attitude_z.append(data[listIndex]['attitude']['z'])
            
            ### gravity
            gravity_x.append(data[listIndex]['gravity']['x'])
            gravity_y.append(data[listIndex]['gravity']['y'])
            gravity_z.append(data[listIndex]['gravity']['z'])

            
#         print(index, len(data))
        if index % 500 == 0:
            print(index, len(acc_x), "processing ---> ", target_arr[index].split('/')[9], fileNameDiagMap.get(target_arr[index].split('/')[9]))
        
        
        if (len(acc_x) < 1024):
            print("Pass. length is short.")
            continue
        
        if str(fileNameDiagMap.get(target_arr[index].split('/')[9])) == 'nan':
            print("identification fail. pass.")
            continue
                  
        
        
        while True:
            if w_end > len(acc_x):
                break

#             print(w_start, w_end)
            
            for i in range(w_start, w_end):
                ############# write acc_x #############
                file_acc_x.write(str(data[i]['userAcceleration']['x']))
                file_acc_x.write(' ')
                
                file_rotation_x.write(str(data[i]['rotationRate']['x']))
                file_rotation_x.write(' ')
                
                file_attitude_x.write(str(data[i]['attitude']['x']))
                file_attitude_x.write(' ')
                
                file_gravity_x.write(str(data[i]['gravity']['x']))
                file_gravity_x.write(' ')
                                
                
                ############# write acc_y #############
                file_acc_y.write(str(data[i]['userAcceleration']['y']))
                file_acc_y.write(' ')
                
                file_rotation_y.write(str(data[i]['rotationRate']['y']))
                file_rotation_y.write(' ')
                
                file_attitude_y.write(str(data[i]['attitude']['y']))
                file_attitude_y.write(' ')
                
                file_gravity_y.write(str(data[i]['gravity']['y']))
                file_gravity_y.write(' ')
                
                
                ############# write acc_z #############
                file_acc_z.write(str(data[i]['userAcceleration']['z']))
                file_acc_z.write(' ')
                
                
                file_rotation_z.write(str(data[i]['rotationRate']['z']))
                file_rotation_z.write(' ')
                
                file_attitude_z.write(str(data[i]['attitude']['z']))
                file_attitude_z.write(' ')
                
                file_gravity_z.write(str(data[i]['gravity']['z']))
                file_gravity_z.write(' ')
                
                
            if str(fileNameDiagMap.get(target_arr[index].split('/')[9])) == 'True':
                file_y_label.write(str(1))
            else:
                file_y_label.write(str(0))
            
            
            
            file_acc_x.write('\n')
            file_acc_y.write('\n')
            file_acc_z.write('\n')
            
            file_rotation_x.write('\n')
            file_rotation_y.write('\n')
            file_rotation_z.write('\n')
            
            file_attitude_x.write('\n')
            file_attitude_y.write('\n')
            file_attitude_z.write('\n')
            
            file_gravity_x.write('\n')
            file_gravity_y.write('\n')
            file_gravity_z.write('\n')
            
            
            
            file_y_label.write('\n')
            file_mapping_label.write(target_arr[index])
            file_mapping_label.write('\n')
        
      
            
            w_start = w_start + (window_size / 2)
            w_end = w_start + window_size
            
        w_start = 0
        w_end = w_start + window_size
    


######### write done. finalize data files
file_acc_x.close()
file_acc_y.close()
file_acc_z.close()

file_rotation_x.close()
file_rotation_y.close()
file_rotation_z.close()

file_attitude_x.close()
file_attitude_y.close()
file_attitude_z.close()

file_gravity_x.close()
file_gravity_y.close()
file_gravity_z.close()


file_y_label.close()
file_mapping_label.close()

print("\nDataset generation done.")


```
