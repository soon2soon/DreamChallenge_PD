```python
import os
import pandas as pd

# dataset path
train_path = "/home/soon2soon/Notebooks/DreamChallenge/sc2/data/train/"
test_path = "/home/soon2soon/Notebooks/DreamChallenge/sc2/data/test/"


id_fileHandleId_train = "/home/soon2soon/Notebooks/DreamChallenge/sc2/Job-48759725444146931342720587_train.csv"
id_fileHandleId_test = "/home/soon2soon/Notebooks/DreamChallenge/sc2/Job-48759775293593331375175371_test.csv"


id_table_train =  pd.read_csv(id_fileHandleId_train)
id_table_train = id_table_train.drop(["ROW_ID", "ROW_VERSION"], axis=1)

id_table_test =  pd.read_csv(id_fileHandleId_test)
id_table_test = id_table_test.drop(["ROW_ID", "ROW_VERSION"], axis=1)

# print(id_table_train)
```

```python
## dataFileHandleId <-> true file path mapper


# path = train_path

i = 0


path_arr = [train_path, test_path]

# print(path_arr)

for path in path_arr:
    if path == train_path:
        mode = 'train'
    else:
        mode = 'test'

    writefile = open(path + "id_fileMapping_"+mode+".txt","w") 

    writefile.write('\"'+"dataFileHandleId"+'\"'+','+ '\"' + "FILE_NAME" + '\"' +','+ '\"'+ "FILE_PATH"+ '\"')
    writefile.write('\n')

    for subdir, dirs, files in os.walk(path):
        for file in files:
    #         print os.path.join(subdir, file)
            filepath = subdir + os.sep + file
    #         print(filepath)

            if ".tsv" in filepath:
                i = i + 1
                string = '\"'+filepath.split('/')[9]+'\"' + "," +'\"'+filepath.split('/')[10]+'\"' + "," +'\"' + filepath +'\"'
                print(i,string)
                writefile.write(string+"\n")

    writefile.close()
    print(mode, "file write done")


```

```python
## now fileHandleID and true file path is linked.

id_filePath_mapping_train = train_path + "id_fileMapping_train.txt"
id_filePath_mapping_test = test_path + "id_fileMapping_test.txt"

id_file_mapping_train = pd.read_csv(id_filePath_mapping_train)
id_file_mapping_test = pd.read_csv(id_filePath_mapping_test)


data_path = "/home/soon2soon/Notebooks/DreamChallenge/sc2/data/"

# print(id_file_mapping_train)


merged_table_train = pd.merge(id_table_train, id_file_mapping_train, on="dataFileHandleId")
merged_table_test = pd.merge(id_table_test, id_file_mapping_test, on="dataFileHandleId")


# print(merged_table_train)
print(merged_table_test)



merged_table_train.to_csv(data_path + "id_file_mapping_train.csv", sep=',')
merged_table_test.to_csv(data_path + "id_file_mapping_test.csv", sep=',')
```
